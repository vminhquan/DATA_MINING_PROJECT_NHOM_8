# Cell 1: Import thư viện và thiết lập hiển thị

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")
plt.rcParams.update({'font.size': 12})
```

# Cell 2: Tổng hợp kết quả Phân lớp (Supervised Classification)

```python
clf_data = {
    'Model': ['Naive Bayes', 'Logistic Regression', 'Linear SVC'],
    'F1-Macro Score': [0.65, 0.76, 0.79]
}
df_clf = pd.DataFrame(clf_data)

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Model', y='F1-Macro Score', data=df_clf, palette='Blues_d')
plt.title('So sánh hiệu năng các mô hình Phân lớp Cảm xúc')
plt.ylim(0, 1)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')
plt.show()
```

# Cell 3: Tổng hợp kết quả Hồi quy (Regression)

```python
reg_data = {
    'Model': ['Baseline (Ridge)', 'Strong Model (Random Forest)'],
    'MAE': [0.85, 0.62],
    'RMSE': [1.12, 0.89]
}
df_reg = pd.DataFrame(reg_data)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x='Model', y='MAE', data=df_reg, ax=axes[0], palette='Oranges_d')
axes[0].set_title('So sánh Sai số tuyệt đối trung bình (MAE)')

sns.barplot(x='Model', y='RMSE', data=df_reg, ax=axes[1], palette='Reds_d')
axes[1].set_title('So sánh Căn bậc hai sai số bình phương (RMSE)')

plt.tight_layout()
plt.show()
```
